```python
import cv2 as cv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv.imread('badmos.jpeg')
plt.imshow(image)
```

```python
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
image_inv=255-image
cv.imwrite('inv_badmos.jpeg',image_inv)
plt.imshow(image_inv)
```

```python
image2=cv.imread('badmos.jpeg')
plt.subplot(131)
plt.imshow(image2)
gr=cv.cvtColor(image2,cv.COLOR_BGR2GRAY)
eq=cv.equalizeHist(gr)
plt.subplot(132)
cv.imwrite('equalised.jpeg',eq)
plt.imshow(eq)
cr_img=image2[100:120,100:120]
plt.subplot(133)
plt.imshow(cr_img)
```
